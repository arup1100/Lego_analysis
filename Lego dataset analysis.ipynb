#### What percentage of all licensed sets ever released were star wars theme


```python
#joining both the data sets to get one view of all data
df_merged = pd.merge(df,parent_theme,left_on = 'parent_theme',right_on='name')
df_merged
```

```python
#drop the name_y column as it appear in both tables
df_merged = df_merged.drop(columns=['name_y'])
df_merged
```

```python
licensed = df_merged[df_merged['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed
```

```python
#check for null values
licensed.isnull().sum()
```

```python
starwars = licensed[licensed['parent_theme'] == 'Star Wars']
starwars.shape
```

```python
#percentage calculation
the_force = int(starwars.shape[0]/licensed.shape[0]*100)
the_force
```

#### In Which Year Star Wars was not the most popular licensed theme(in terms of no of sets released that year)

```python
df_merged
```

```python
popular_licensed = df_merged[df_merged['is_licensed']]
popular_licensed
```

```python
popular = popular_licensed.groupby(['year','parent_theme']).sum().reset_index().sort_values(by ='year',ascending = False)
# popular = popular[popular['parent_theme'] == 'Star Wars']
popular
```

```python
popular_new = popular.sort_values('is_licensed',ascending=False).drop_duplicates('year')
popular_new
```

```python
popular_new = popular_new[popular_new['parent_theme'] != 'Star Wars']
popular_new
```
